```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# رابط الصفحة الرئيسية للـ reviews
url = "https://www.trustpilot.com/review/www.airalgerie.dz"

headers = {
    "User-Agent": "Mozilla/5.0"
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

# استخراج جميع التعليقات
reviews = soup.select('p[data-service-review-text-typography="true"]')

data = []

for review in reviews:
    text = review.get_text(strip=True)
    data.append(text)

# تحويل إلى DataFrame
df = pd.DataFrame(data, columns=["commentaire"])

# حفظ CSV
df.to_csv("air_algerie_trustpilot_reviews.csv", index=False, encoding="utf-8")

print("✅ Scraping terminé")
print("Nombre de commentaires extraits :", len(df))
```
